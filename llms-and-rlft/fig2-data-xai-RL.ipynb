```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

```python
def get_fig_data(input_data):
    bins = 10
    data = input_data
    breaks = np.linspace(np.min(data), np.max(data), bins)
    counts = {}
    for i in range(len(breaks) - 1):
        left, right = breaks[i], breaks[i+1]
        label = f"({left:.2f}, {right:.2f}]"
        count = 0
        for val in data:
            if val > left and val <= right:
                count += 1
        counts.update({label: count})
    print(counts)
    pd.cut(pd.Series(data), bins=breaks, precision=2).value_counts()
    return data
```

```python
def make_2data_fig(input_data1,input_label1,input_data2,input_label2,title,save_path=None):
    data1 = input_data1
    label1 = input_label1
    data2 = input_data2
    label2 = input_label2
    fig_title =title # "QED Scores",
    
    kwargs = {
        "bins": 40,
        "histtype": "stepfilled",
        "alpha": 0.5
    }
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.hist(data1, label=label1, **kwargs)
    ax.hist(data2, label=label2, **kwargs)
    
    #ax.set_xlabel("Value", fontsize=14)
    #ax.set_ylabel("Frequency", fontsize=14)
    #ax.legend(fontsize=12)

    # Ticks and grid
    ax.tick_params(axis='both', labelsize=16)
    #ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

    ax.set_title(fig_title, fontsize=18, weight='bold')
    ax.legend(fontsize=15,loc="upper left")
    #plt.savefig('qed-1.svg', dpi=300,format='svg')
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', format='jpg')
```

```python
make_2data_fig(qed_scores_gcpo,'CPO@Dense',qed_scores_gxcpo,'XCPO@Dense', "QED Scores",'dense-qed.jpg')
```

```python
make_2data_fig(sa_scores_gcpo,'CPO@Dense',sa_scores_gxcpo,'XCPO@Dense', "SA Scores",'dense-sa.jpg')
```

```python
def cal_nbits_vect_smi(smis, nbit):
    radius = 2
    mol = Chem.MolFromSmiles(smis)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nbit)
    fp_list = list(fp.ToBitString())
    fp_list = [int(bit) for bit in fp_list]
    #fp_list.extend([0] * (nbit - len(fp_list)))
    return fp_list

def cal_nbits_vect(mols, nbit):
    radius = 2
    fp = AllChem.GetMorganFingerprintAsBitVect(mols, radius, nBits=nbit)
    fp_list = list(fp.ToBitString())
    fp_list = [int(bit) for bit in fp_list]
    #fp_list.extend([0] * (nbit - len(fp_list)))
    return fp_list

def get_1024fp_set(mol_set):
    n_bits = 1024
    fp_set =[]
    for item in range(len(mol_set)):
        mol_ = mol_set[item]
        fp_ = cal_nbits_vect(mol_,n_bits)
        fp_set.append(fp_)
    return np.asarray(fp_set)
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
fp_gcpo = get_1024fp_set(mols_gcpo)
```

```python
fp_gxcpo = get_1024fp_set(mols_gxcpo)
```

```python
fp_qcpo = get_1024fp_set(mols_qcpo)
```

```python
fp_qxcpo = get_1024fp_set(mols_qxcpo)
```

```python
pca_ = PCA(n_components=20)
```

```python
pca_gcpo = pca_.fit_transform(fp_gcpo)
```

```python
pca_gxcpo = pca_.fit_transform(fp_gxcpo)
```

```python
pca_qcpo = pca_.fit_transform(fp_qcpo)
```

```python
pca_qxcpo = pca_.fit_transform(fp_qxcpo)
```

```python
vec_dense = np.concatenate((pca_gcpo, pca_gxcpo), axis=0)
```

```python
vec_moe = np.concatenate((pca_qcpo, pca_qxcpo), axis=0)
```

```python
tsne = TSNE(min_grad_norm=1e-3,init='pca',method='exact',angle=1,early_exaggeration=5,n_iter=1000)
```

```python
t_dense = tsne.fit_transform(vec_dense)
```

```python
t_moe = tsne.fit_transform(vec_moe)
```

```python
len(pca_gcpo)
```

```python
gcpo_vec = []
gxcpo_vec =[]
for item_ in range(len(t_dense)):
    data = t_dense[item_]
    if item_ < len(pca_gcpo):
        gcpo_vec.append(data)
    else:
        gxcpo_vec.append(data)
```

```python
qcpo_vec = []
qxcpo_vec =[]
for item_ in range(len(t_moe)):
    data = t_moe[item_]
    if item_ < len(pca_gcpo):
        qcpo_vec.append(data)
    else:
        qxcpo_vec.append(data)
```

```python
gcpo_vec = np.asarray(gcpo_vec)
gxcpo_vec = np.asarray(gxcpo_vec)
qcpo_vec = np.asarray(qcpo_vec)
qxcpo_vec = np.asarray(qxcpo_vec)
```

```python
gx1 = gcpo_vec.T[0]
gy1 = gcpo_vec.T[1]
gx2 = gxcpo_vec.T[0]
gy2 = gxcpo_vec.T[1]
qx1 = qcpo_vec.T[0]
qy1 = qcpo_vec.T[1]
qx2 = qxcpo_vec.T[0]
qy2 = qxcpo_vec.T[1]
```
